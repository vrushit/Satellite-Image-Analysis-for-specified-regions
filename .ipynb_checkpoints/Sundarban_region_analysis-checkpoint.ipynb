#Exploring Sundarban Wildlife Santuary Land coverage

```python
#Importing libraries which majorly include earthpy and glob which are used to handle GeoJSON format and TIFF format
from glob import glob

import earthpy as et
import earthpy.spatial as es
import earthpy.plot as ep

import rasterio as rio
from rasterio.plot import plotting_extent
from rasterio.plot import show
from rasterio.plot import reshape_as_raster, reshape_as_image

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

from scipy.io import loadmat
from sklearn.metrics import classification_report, accuracy_score

import plotly.graph_objects as go

#Ignoring the unwanted variables and atmospheric coefficient
np.seterr(divide='ignore', invalid='ignore')
```

```python
from google.colab import drive
drive.mount('/content/drive')

%cd /content/drive/My Drive/sat_data

```

```python

#Using sundarban data in tiff format
sundarban_data_tiffs = glob("/content/drive/MyDrive/sat_data/*B?*.tiff")
sundarban_data_tiffs.sort()
sundarban_data_tiffs

```

```python


l = []
for i in sundarban_data_tiffs:
  with rio.open(i, 'r') as f:
    l.append(f.read(1))

```

```python
arr_st = np.stack(l)
```

```python
print(f'Height: {arr_st.shape[1]}\nWidth: {arr_st.shape[2]}\nBands: {arr_st.shape[0]}')
```

```python


ep.plot_bands(arr_st, cmap = 'gist_earth', figsize = (20, 12), cols = 6, cbar = False)
plt.show()
# Band representing the different color wavelength
```

```python
ndvi = es.normalized_diff(arr_st[7], arr_st[3])

ep.plot_bands(ndvi, cmap="RdYlGn", cols=1, vmin=-1, vmax=1, figsize=(8, 10))

plt.show()
```

# As We can see the vegatation coverage for sundarban is very vast in terms of NDVI Index which even reflects on the tiff image.
